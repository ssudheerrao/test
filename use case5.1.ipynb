```python
df.drop('id',axis=1,inplace=True)
df
```

```python
df_num=df.select_dtypes(include=['int64','float64']).copy()
df_num
```

```python
df_cat=df.select_dtypes(include=['object']).copy()
df_cat
```

```python
df['education'].value_counts()
```

```python

# Import label encoder for single column
from sklearn import preprocessing 
  
# label_encoder object knows how to understand word labels. 
label_encoder = preprocessing.LabelEncoder() 
  
# Encode labels in column 'species'. 
df_cat['difficulty_level']= label_encoder.fit_transform(df_cat['difficulty_level']) 
  
df_cat

## Label encoder Multiple columns

columns=['difficulty_level','education']
from sklearn.preprocessing import LabelEncoder
le_dict = { col: LabelEncoder() for col in columns }
for col in columns:
   df_cat[col]= le_dict[col].fit_transform(df_cat[col])

df_cat
```

```python
df_cat.info()
```

```python
Label_nums = {"difficulty_level":     {"easy": 1, "intermediate": 2, "hard":3, "vary hard":4},
                "education": {"No Qualification": 1, "Matriculation": 2, "High School Diploma": 3, "Bachelors": 4,
                                  "Masters": 5}}
```

```python
df_cat=df_cat.replace(Label_nums)
df_cat
```

```python
pd.get_dummies(df['is_handicapped'],drop_first=True)
```

```python
columns_to_dummy = ['program_id','program_type','test_type',
                   'gender','is_handicapped']
df_cat = pd.get_dummies(df_cat, columns=columns_to_dummy,drop_first=True)
```

```python
df_final=pd.concat([df_num,df_cat],axis=1)
df_final
```

```python
df_final.isnull().sum()
```

```python
df_final['age']=df_final['age'].fillna(df_final['age'].mean())
df_final['trainee_engagement_rating']=df_final['trainee_engagement_rating'].fillna(df_final['trainee_engagement_rating'].mean())
df_final
```

```python
X=df_final.iloc[:,:-1]
y=df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.01,random_state=40)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
```

```python
rf.score(x_train, y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,f1_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
print('f1 score',f1_score(y_test,y_pred))
```

```python
from sklearn.metrics import roc_curve, auc
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(false_positive_rate, true_positive_rate)
roc_auc
```
